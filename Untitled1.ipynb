# Sentence Tokenization

```python
# Install and import
import nltk
nltk.download('punkt')
nltk.download('punkt_tab')

from nltk.tokenize import sent_tokenize

text = """Natural Language Processing is a field of AI.
It helps computers understand human language.
It is widely used in chatbots, translation, and search engines."""

sentences = sent_tokenize(text)

print("Sentences:")
for s in sentences:
    print(s)
```

# Removing Stop words

```python
import nltk
nltk.download('punkt')
nltk.download('stopwords')

from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize

text = "NLP is an important field of artificial intelligence."
tokens = word_tokenize(text)

stop_words = set(stopwords.words('english'))
filtered_tokens = [word for word in tokens if word.lower() not in stop_words]

print("Filtered Tokens:")
print(filtered_tokens)
```

# Stemming

```python
import nltk
from nltk.stem import PorterStemmer

stemmer = PorterStemmer()

words = ["running", "jumps", "easily", "studies"]
stems = [stemmer.stem(word) for word in words]

print("Stems:")
print(stems)
```

# Parsing

```python
import nltk
from nltk import CFG
from IPython.display import display

grammar = CFG.fromstring("""
S -> NP VP
NP -> Det N
VP -> V NP
Det -> 'the'
N -> 'cat' | 'dog'
V -> 'chased'
""")

parser = nltk.ChartParser(grammar)

sentence = "the cat chased the dog".split()

for tree in parser.parse(sentence):
    print(tree)
    display(tree)
```

# Named Entity Recognition

```python
import nltk
nltk.download('punkt')
nltk.download('averaged_perceptron_tagger')
nltk.download('maxent_ne_chunker')
nltk.download('words')

text = "Barack Obama was born in Hawaii."

tokens = nltk.word_tokenize(text)
ner_tree = nltk.ne_chunk(nltk.pos_tag(tokens))

print("Named Entities:")

for subtree in ner_tree:
    if hasattr(subtree, 'label'):
        entity = " ".join([token for token, pos in subtree.leaves()])
        print(entity, "->", subtree.label())
```

# Word Sense Disambiguation

```python
import nltk
nltk.download('punkt')
nltk.download('wordnet')

from nltk.corpus import wordnet as wn
from nltk.tokenize import word_tokenize

sentence = "He sat on the bank of the river."
word = "bank"

tokens = set(word_tokenize(sentence.lower()))

best_sense = None
max_overlap = 0

for sense in wn.synsets(word, pos=wn.NOUN):
    definition = set(word_tokenize(sense.definition().lower()))
    overlap = len(tokens.intersection(definition))

    if overlap > max_overlap:
        max_overlap = overlap
        best_sense = sense

print("Correct Sense:", best_sense)
print("Definition:", best_sense.definition())
```

# Common Regex Functions

```python
import re

text = "Natural Language Processing with Python"

pattern = r"Language"

result = re.search(pattern, text)

if result:
    print("Match found:", result.group())
    print("Start index:", result.start())
else:
    print("No match found")
```

```python
import re

text = "Natural Language Processing"
pattern = r"Natural"

result = re.match(pattern, text)

if result:
    print("Match found at the beginning:", result.group())
else:
    print("No match at the beginning")
```

```python
import re

text = "NLP stands for Natural Language Processing."

pattern = r"\b[A-Z]{2,}\b"   # words with 2+ uppercase letters

matches = re.findall(pattern, text)

print("All matches:", matches)
```

```python
import re

text = "I love NLP. NLP is interesting."

pattern = r"NLP"
replacement = "Natural Language Processing"

new_text = re.sub(pattern, replacement, text)

print("Updated text:")
print(new_text)
```

```python
import re

text = "Email me at student123@gmail.com or test_user@yahoo.com"

# compile pattern for email detection
email_pattern = re.compile(r"\b[\w\.-]+@[\w\.-]+\.\w+\b")

emails = email_pattern.findall(text)

print("Emails found:")
print(emails)
```
